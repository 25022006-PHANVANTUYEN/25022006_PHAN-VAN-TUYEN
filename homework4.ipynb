```python
#a6 nhap a,b tinh tong snt trong doan (a,b)
from math import isqrt
def snt(i):
    if i < 2:
        return False
    for j in range(2, isqrt(i) + 1):
        if i % j == 0:
            return False
    return True
a, b = map(int, input().split())
while a >= b:
    a, b = map(int, input('nhap lai:').split())
s = 0
for i in range(a, b+1):
    if snt(i):
        s = s + i
print(s)
```

```python
#a8 Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
# đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.
n = int(input())
count = 0
dap = n
flag = 0


while flag == 0:
    flag = 1
    count += 1
    k = str(n)
    k = k[::-1]
    k = int(k)
    dap += k
    dap = str(dap)
    for i in range(1, len(dap) // 2):
        if dap[i] != dap[len(dap) - i - 1]:
            flag = 0
            break
    dap = str(dap)

print('so lan =', count)
print('so doi xung =', dap)
     
```

```python
#a10 : nhap vao so nguyen duong n, tim 1<x<=n co do dai day collatz lon nhat
# va in ra (x L)

def len(s):
    L = 0
    while (s != 1):
        L += 1
        if (s % 2 == 0):
            s = s / 2
        else:
            s = 3 * s + 1

    return L

n = int(input())
L = 1
x = 1
for i in range(1, n + 1):
    l = len(i)
    if(L < l):
        L = l
        x = i

print(x, L)
```

```python
# a13 
# nhập vào 2 số kiểm tra xem 2 số xem có thân thiết với nhau không

def tong(k):
    s = 0
    for i in range(1, k):
        if (k % i == 0):
            s += i
    return s

a, b = map(int, input().split())

if ((a == tong(b)) and (b == tong(a))):
    print('true')
else:
    print('false')
```

```python
#w4a14
#nhap 2 so m,n tim uoc chung lon nhat

m,n = map(int,input('nhap (m,n)=').split())
def gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m
print(gcd(m,n))
```
